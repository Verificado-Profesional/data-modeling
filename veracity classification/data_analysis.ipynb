```python
import pandas as pd
import matplotlib.pyplot as plt
from unidecode import unidecode
```

### Data

```python
data = pd.read_csv("data/data.csv")
data.shape
```

```python
tf_count = data["clasificacion"].value_counts()
tf_count
```

```python
plt.figure(figsize=(10,10))
plt.pie(tf_count, labels=["true", "fake"],autopct="%1.f%%", colors=["skyblue", "crimson"], wedgeprops={'linewidth': 0.2, 'edgecolor': 'black'})

plt.title("Cantidad de noticias verdaderas y falsas en dataset procesado")

plt.show()
```

### Data Chequeado

```python
chequeado = pd.read_csv("data/chequeado.csv")
chequeado.shape
```

```python
tf_count = chequeado["clasificacion"].value_counts()
tf_count
```

```python
plt.figure(figsize=(10,10))
plt.pie(tf_count, labels=["fake","true"],autopct="%1.f%%", colors=["crimson","skyblue"], wedgeprops={'linewidth': 0.2, 'edgecolor': 'black'})

plt.title("Cantidad de noticias Verdaderas y Falsas en dataset de Chequeado")
```

### Integracion

```python
data_join = pd.concat([data, chequeado])
data_join.shape
```

```python
tf_count = data_join["clasificacion"].value_counts()
tf_count
```

```python
plt.figure(figsize=(10,10))
plt.pie(tf_count, labels=["true", "fake"],autopct="%1.f%%", colors=["skyblue", "crimson"], wedgeprops={'linewidth': 0.2, 'edgecolor': 'black'})

plt.title("Cantidad de noticias verdaderas y falsas en Dataset Integrado")

plt.show()
```
